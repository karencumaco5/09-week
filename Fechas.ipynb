```python
import datetime
import pandas as pd

# Assuming 'df' is your DataFrame and 'fechas' is a column with dates
# Replace this with your actual DataFrame if needed
# Example DataFrame
data = {'fechas': ['2023-01-15', '2023-02-20', '2023-03-10']}
df = pd.DataFrame(data)
df['fechas'] = pd.to_datetime(df['fechas'])

fecha_referencia = pd.to_datetime('2023-01-01')
df['días_transcurridos'] = (df['fechas'] - fecha_referencia).dt.days
df  # Display the DataFrame
(df['fechas'] - fecha_referencia).dt.days  # Calculate and display the days
df  # Display the DataFrame again
```

Jerarquía de Fechas

Descripción

Permite organizar datos temporales en diferentes niveles: año, trimestre, mes, semana, día. Esto facilita el ánalisis agregado y la visualización en reportes y dashboards.

Aplicabilidad

Ideal para:



*   Dashboards interactivos
*   Reportes mensuales, trimestrales, anuales

*   Pronósticos









```python
import numpy as np
import pandas as pd

# Crear un Dataframe con ventas diarias
df = pd.DataFrame({
    'fechas': pd.date_range(start='2023-01-01', periods=90, freq='D'),
    'ventas': np.random.randint(100, 1000, size=90)
})

# Agrupar por mes
df['mes'] = df['fechas'].dt.to_period('M')
ventas_mensuales = df.groupby('mes')['ventas'].sum()

# To get the total sales, you should use:
total_ventas = ventas_mensuales.sum()  # Sum the values in the 'ventas' column of the Series

# Print or display the results
print("Ventas mensuales:\n", ventas_mensuales)
print("\nTotal de ventas:", total_ventas)
```

Actividad



1.   Agrupa un dataset por trimestre y calcula estadísticas descriptivas.



```python
df['trimestre'] = df['fechas'].dt.to_period('Q') # Changed 'fecha' to 'fechas'
estadísticas_trimestrales = df.groupby('trimestre')['ventas'].describe()
estadísticas_trimestrales
```



1.   Visualiza las ventas promedio por mes.



```python
import matplotlib.pyplot as plt

ventas_mensuales_promedio = df.groupby('mes')['ventas'].mean()
ventas_mensuales_promedio.plot(kind='bar', title='Ventas promedio por Mes')
plt.ylabel('Ventas promedio')
plt.xlabel('Mes')
plt.tight_layout()
plt.show()
```
